```python
import pandas as pd
```

```python
df = pd.read_csv('data/fuzzy_test_class_only.csv', index_col=[0])
#df.reset_index(drop=True, inplace=True)
df.tail()
```

```python
def func(x):
    try :
        x=eval(x)
        res = pd.Series([
            x[0][0], x[0][1] , x[1][0], x[1][1], x[1][2]
            ])
        return res
    except : return x

df[['match', 'score', 'dyad', 'session', 'time_begin']] = df.Match.apply(func)
df.drop(columns=['Match'], inplace=True)
df.head()
```

```python
df_int = df[['dyad', 'session', 'time_begin','match', 'score', 'SD', 'QE', 'SV', 'PR', 'HD']].values
df_int = pd.DataFrame(df_int, columns=['dyad', 'session', 'time_begin','match', 'score', 'SD', 'QE', 'SV', 'PR', 'HD'])
df_int['id'] = df_int.dyad.astype(str) + ' - ' + df_int.session.astype(str) + ' - ' + pd.to_datetime(df_int.time_begin, format='%H:%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
df_int['idd'] = df_int.match.astype(str) + ' - ' + pd.to_datetime(df_int.time_begin, format='%H:%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
df_int.head()
```

```python
full = pd.read_csv("data/transcripts/whole_data_2021_2016.csv", index_col=[0])
full['id'] = full.Dyad.astype(str) + ' - ' + full.Session.astype(str) + ' - ' + pd.to_datetime(full.Time_begin, format='%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
full['id1'] = full.P1.astype(str)+ ' - ' + pd.to_datetime(full.Time_begin, format='%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
full['id2'] = full.P2.astype(str)+ ' - ' + pd.to_datetime(full.Time_begin, format='%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
full.head()
```
