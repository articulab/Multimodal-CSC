```python
import pandas as pd
```

```python
#import data from matching
fuzzy_raw = pd.read_csv('data/fuzzy_test_class_only.csv', index_col=[0])
fuzzy_raw.tail()
```

```python
#Breaking the 'Match' column into columns ['match', 'score', 'dyad', 'session', 'time_begin']
#putting the result in fuzzy_processed

def func(x):
    """
    To be used in an apply()
    """
    try :
        x=eval(x)
        res = pd.Series([
            x[0][0], x[0][1] , x[1][0], x[1][1], x[1][2]
            ])
        return res
    except : return x

fuzzy_processed = pd.DataFrame(
    fuzzy_raw.values,
    columns=fuzzy_raw.columns
    )

fuzzy_processed[['match', 'score', 'dyad', 'session', 'time_begin']] = fuzzy_raw.Match.apply(func).values
fuzzy_processed.drop(columns=['Match'], inplace=True)
fuzzy_processed.head()
```

```python
#df_inter to be merged with the full dataset
df_inter = pd.DataFrame(fuzzy_processed[['score', 'SD', 'QE', 'SV', 'PR', 'HD']].values, columns=['score', 'SD', 'QE', 'SV', 'PR', 'HD'])

#creating an 'id' column and putting it to index
df_inter['id'] = fuzzy_processed.match.astype(str) + ' - ' + fuzzy_processed.dyad.astype(str) + ' - ' + fuzzy_processed.session.astype(str) + ' - ' + pd.to_datetime(fuzzy_processed.time_begin, format='%H:%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
df_inter.set_index('id', drop=True, inplace=True)

df_inter.head()
```

```python
#importing full dataset into 'full'
full = pd.read_csv("data/aligned_transcript_nolabel.csv", index_col=[0])

#reproducing the id column
full['id1'] = full.P1.astype(str)+ ' - ' + full.Dyad.astype(str) + ' - ' + full.Session.astype(str) + ' - ' + pd.to_datetime(full.Time_begin_ts, format='%Y-%m-%d %H:%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
full['id2'] = full.P2.astype(str)+ ' - ' + full.Dyad.astype(str) + ' - ' + full.Session.astype(str) + ' - ' + pd.to_datetime(full.Time_begin_ts, format='%Y-%m-%d %H:%M:%S.%f').dt.strftime("%H:%M:%S.%f").str.split('.', expand=True)[0]
full['id'] = full[['id1', 'id2']].apply(
    lambda x : x.id1 if x.id1[:3] != 'nan' else x.id2, #empty utt always starts with 'nan'
    axis=1
    )
full.drop(columns=['id1', 'id2'], inplace=True)

full.head()
```

```python
#safety checks
print(len(df_int))
print(len(set(df_int.id)))
print(len(set(full.id).intersection(set(df_int.id))))
#at least the id column works but some matches have possibly been assigned to multiple rows or the inverse (whatever - it can be negleted)
```

```python
#make result
result = full.merge(
    df_inter[['score','SD','QE','SV','PR','HD']],
    how='left',
    on='id'
    )
result.drop(columns=['id'], inplace=True)

#############
## Save it ##
#############

#result.to_csv('result.csv')

result.head()
# to see non empty cat run
#result.loc[~result.score.isna()]
```
